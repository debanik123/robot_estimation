```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gassuan(x, mu, var):
    return (1/np.sqrt(2*np.pi*var))*np.exp(-(x-mu)**2/var)
```

```python
gassuan(10, 4, 2)
```

```python
x_values = np.linspace(-1, 1, 120)
for mu, sig in [(-1, 1), (0, 2), (2, 3)]:
    plt.plot(x_values, gassuan(x_values, mu, sig))

plt.show()
```

```python
def update(m1, v1, m2, v2):
    new_m = (v2*m1 + v1*m2)/(v1+v2)
    new_v = 1/((1/v1)+(1/v2))
    return [new_m, new_v]

print(update(10, 5, 20, 10))
```

```python
update_param = update(10, 5, 20, 10)
```

```python
new_mu = update_param[0]
new_var = update_param[1]

x = np.array(np.arange(0, 20, 0.1))
#print(gassuan(x, new_mu, new_var))

plt.plot(x, gassuan(x, new_mu, new_var))
plt.show()
```

```python
new_mu = update_param[0]
new_var = update_param[1]

x = np.arange(0, 20, 0.1)
g = []

for i in x:
    g.append(gassuan(i, new_mu, new_var))

plt.plot(x, g)
plt.show()
```

```python
def predict(m1,v1, m2, v2):
    new_m = m1+m2
    new_v = v1+v2
    return [new_m, new_v]

print(predict(10, 5, 20, 10))
```

```python
predict_param = predict(10, 5, 20, 10)
```

```python
new_mu = predict_param[0]
new_var = predict_param[1]

x = np.array(np.arange(0, 60, 0.1))
#print(gassuan(x, new_mu, new_var))

plt.plot(x, gassuan(x, new_mu, new_var))
plt.show()
```

```python
measurement = [5., 6., 7., 9., 10. ]
motion = [1., 1., 2., 1., 1.]
measument_var = 4
measurement_m = 2
mu = 0
var = 1000

for n in range(len(measurement)):
    [mu, var] = update(mu, var, measurement[n], measument_var)
    print('update:', [mu, var])
    [mu, var] = predict(mu, var, motion[n], measument_var)
    print('predict:', [mu, var])
    
```
